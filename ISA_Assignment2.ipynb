```python
# Load the CSV file
file_path = 'train.csv'
data = pd.read_csv(file_path)
```

#a) Look for the missing values in all the columns and either impute them (replace with mean, median, or mode) or drop them. Justify your action for this task.

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
```

#### Imputing Missing Values

- **Numerical Columns**: I will use the mean or median for imputation to maintain the integrity of the dataset. The median will be my choice for skewed data because it is less impacted by outliers.

- **Categorical Columns**: I will fill missing values with the mode to ensure that the most frequently occurring category is preserved, thus maintaining the categorical structure of the data.

### Dropping Columns/Rows

- **Dropping Columns**: I will consider removing columns with more than 20% missing values, as these columns may lack sufficient information and could complicate the analysis.

- **Dropping Rows**: If certain rows contain a significant number of missing values, and the remaining rows still represent the dataset well, I will drop those rows to preserve data quality while minimizing information loss.

```python
# Impute or drop missing values
# Example strategy: Impute with mean for numerical columns, mode for categorical
for column in data.columns:
    if data[column].isnull().sum() > 0:  # If there are missing values
        if data[column].dtype == 'object':  # If column is categorical
            # Impute with mode
            data[column].fillna(data[column].mode()[0], inplace=True)
        else:  # If column is numerical
            # Impute with mean
            data[column].fillna(data[column].mean(), inplace=True)
```

```python
# Verify if all missing values are handled
missing_values_after = data.isnull().sum()
print("\nMissing values after imputation:\n", missing_values_after)
```

#b) Remove the units from some of the attributes and only keep the numerical values.

```python
# Function to clean and convert columns with units to numeric
def clean_numeric_column(column):
    return pd.to_numeric(column.str.replace(r'[^0-9.]', '', regex=True), errors='coerce')
```

```python
# List of columns to clean
columns_to_clean = ['Mileage', 'Engine', 'Power', 'New_Price']

# Apply cleaning function to each column
for column in columns_to_clean:
    if column in data.columns:
        data[column] = clean_numeric_column(data[column])

# Display the cleaned dataset
print(data.head())
```

#C) Change the categorical variables (Fuel_Type and Transmission) into numerical one hot encoded value

```python
# Define the categorical features to be one-hot encoded
cf = ['Fuel_Type', 'Transmission']

# Perform one-hot encoding for all specified categorical columns
data = pd.get_dummies(data, columns=cf, drop_first=False)

# Display the updated DataFrame
print(data)
```

#d) Create one more feature and add this column to the dataset

```python
# Calculate the current year
current_year = datetime.now().year

# Create a new feature 'Car_Age' by subtracting 'Year' from the current year
data['Car_Age'] = current_year - data['Year']

# Display the updated DataFrame
print(data)
```

```python
# Check the columns in the DataFrame
print(data.columns)
```

```python
# Selecting relevant columns
selected_data = data[['Name', 'Location', 'Year', 'Mileage', 'Price', 'Car_Age']]
```

```python
# Filtering rows where Mileage is greater than 18
filtered_data = selected_data[selected_data['Mileage'] > 18]
```

```python
# Renaming columns for clarity
renamed_data = filtered_data.rename(columns={
    'Name': 'Car Name',
    'Location': 'City',
    'Year': 'Manufacture Year',
    'Price': 'Selling Price'
})
```

```python
# Adding a new column 'Price_per_Km' (price divided by kilometers driven)
renamed_data['Price_per_Km'] = renamed_data['Selling Price'] / (data['Kilometers_Driven'] + 1)  # +1 to avoid division by zero
```

```python
# Sorting by 'Selling Price' in descending order
arranged_data = renamed_data.sort_values(by='Selling Price', ascending=False)
```

```python
# Summarizing data: Group by 'City' and calculate mean mileage and selling price
summary_data = arranged_data.groupby('City').agg({
    'Mileage': 'mean',
    'Selling Price': 'mean',
    'Car_Age': 'mean',
    'Price_per_Km': 'mean'
}).reset_index()
```

```python
# Display the outputs of each operation
print("Selected Data:")
print(selected_data, "\n")

print("Filtered Data:")
print(filtered_data, "\n")

print("Renamed Data:")
print(renamed_data, "\n")

print("Arranged Data:")
print(arranged_data, "\n")

print("Summary Data:")
print(summary_data)
```
