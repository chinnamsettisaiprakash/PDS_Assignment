```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('Frailty_Data.csv')

# Show the first few rows of the dataset to understand its structure
data.head()
```

```python
# Checking for null values in the dataset
null_values = data.isnull().sum()

# Displaying the result
null_values
```

```python
# Create a scatter plot for Grip Strength vs. Height with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='Height', y='Grip strength', data=data, marker='o', scatter_kws={'color': 'Red'}, line_kws={'color': 'Green'})
plt.title('Grip Strength vs. Height with Regression Line')
plt.xlabel('Height (inches)')
plt.ylabel('Grip Strength (kg)')
plt.grid(True)
plt.show()


```
